```python

dg = df["tempo"]
array = dg.to_numpy()
```

```python
#Média:
def media(v):
  n = len(v)
  soma = 0
  for i in range(n):
      soma = soma + v[i]
  soma = soma / n
  return soma
print("Média:", "%.2f" %media(array))
```

```python
#Mediana:
def mediana(v):
  n = len(v)
  vo = sorted(v)
  if n % 2 == 0:
      ni = int(n/2)
      mediana = (vo[ni - 1] + vo[ni]) / 2
  else:
      ni = int(n/2)
      mediana = vo[ni]
  return mediana
print("Mediana:", mediana(array))
```

```python
#Variância populacional:
def varianciaP(v):
  n = len(array)
  m = media(array)
  soma = 0
  for i in range(n):
      soma = soma + (v[i] - m)**2
  soma = soma / n
  return soma
print("Variância populacional:", "%.3f" % varianciaP(array))
```

```python
#Variância amostral:
def varianciaA(v):
  n = len(array)
  m = media(array)
  soma = 0
  for i in range(n):
      soma = soma + (v[i] - m)**2
  soma = soma / (n-1)
  return soma
print("Variância amostral:", "%.3f" % varianciaA(array))
```

```python
#Desvio padrão populacional
def dpp(v):
  vp = varianciaP(array)
  dpP = math.sqrt(vp)
  return dpP
print("Desvio padrão populacional:", "%.3f" % dpp(array))
```

```python
#Desvio padrão amostral
def dpa(x):
  va = varianciaA(array)
  dpa = math.sqrt(va)
  return dpa
print("Desvio padrão amostral:", "%.3f" % dpa(array))
```

```python
#Coefiente de variação populacional
def cvp(v):
  cvp = (dpp(array)/media(array))*100
  return cvp
print("Coefiente de variação populacional:" "%.2f" % cvp(array), "%")
```

```python
#Coefiente de variação amostral
def cva(v):
  cva = (dpa(array)/media(array))*100
  return cva
print("Coefiente de variação amostral:" "%.2f" % cva(array), "%")
```

```python
#Quartil 1
posicao_mediana = int((n+1)/2)
ordem = sorted(array)
metade1 = ordem[0:posicao_mediana]
q1 = mediana(metade1)
print("Quartil 1: ", q1)
```

```python
#Quartil 3
posicao_mediana = int((n+1)/2)
ordem = sorted(array)
metade2 = ordem[posicao_mediana:]
q3 = mediana(metade2)
print("Quartil 3: ",q3)
```
