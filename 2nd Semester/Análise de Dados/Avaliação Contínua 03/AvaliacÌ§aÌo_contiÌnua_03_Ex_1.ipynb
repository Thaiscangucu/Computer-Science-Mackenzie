```r
#Média
media <- function(lista) {
  soma <- sum(idade)
  comprimento <- length(idade)
  media <- soma/comprimento
  return(media)
}
media(idade)
```

```r
#Mediana
mediana <- function(lista) {
  n <- length(idade)
  sorted_list <- sort(idade)
  if(n %% 2 == 0){
    ni <- n/2
    mediana <- (sorted_list[ni - 1] + sorted_list[ni]) / 2
  }
  else{
    ni <- n/2
    mediana <- sorted_list[ni]
  }
  return(mediana)
  }
mediana(idade)
```

```r
#Variância Populacional
variancia_P <- function(lista) {
  n <- length(lista)
  m <- mean(lista)
  soma_de_quadrados <- sum((lista - m)^2)
  variancia_P <- soma_de_quadrados / n
  return(variancia_P)
}

variancia_P(idade)
```

```r
#Variância amostral:
variancia_A <- function(lista) {
  n <- length(lista)
  m <- mean(lista)
  soma_de_quadrados <- sum((lista - m)^2)
  variancia_A <- soma_de_quadrados / (n-1)
  return(variancia_A)
}

variancia_A(idade)
```

```r
#Desvio Padrão Populacional
dp <- function(lista) {
  vp <- variancia_P(lista)
  dp <- sqrt(vp)
  return(dp)
}

dp(idade)
```

```r
#Desvio Padrão Amostral
dpa <- function(lista) {
  va <- variancia_A(lista)
  dpa <- sqrt(va)
  return(dpa)
}

dpa(idade)
```

```r
#Coefiente De Variação Populacional:
cv <- function(lista) {
  cv <- (dp(idade)/mean(idade))*100
  return(cv)
}

cv(idade)
```

```r
#Coefiente De Variação Amostral:
cva <- function(lista) {
  cva <- (dpa(idade)/mean(idade))*100
  return(cva)
}

cva(idade)
```
