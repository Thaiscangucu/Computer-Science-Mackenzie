```r
#Média
media <- function(lista) {
  soma <- sum(data$tempo)
  comprimento <- length(data$tempo)
  media <- soma/comprimento
  return(media)
}
round(media(data$tempo))
```

```r
#Mediana
mediana <- function(lista) {
  n <- length(data$tempo)
  sorted_list <- sort(data$tempo)
  if(n %% 2 == 0){
    ni <- n/2
    mediana <- (sorted_list[ni - 1] + sorted_list[ni]) / 2
  }
  else{
    ni <- n/2
    mediana <- sorted_list[ni]
  }
  return(mediana)
  }
round(mediana(data$tempo))
```

```r
#Variância Populacional
variancia_P <- function(lista) {
  n <- length(data$tempo)
  m <- mean(data$tempo)
  soma_de_quadrados <- sum((data$tempo - m)^2)
  variancia_P <- soma_de_quadrados / n
  return(variancia_P)
}

round(variancia_P(data$tempo))
```

```r
#Variância amostral:
variancia_A <- function(lista) {
  n <- length(data$tempo)
  m <- mean(data$tempo)
  soma_de_quadrados <- sum((data$tempo - m)^2)
  variancia_A <- soma_de_quadrados / (n-1)
  return(variancia_A)
}

round(variancia_A(data$tempo))
```

```r
#Desvio Padrão Populacional
dp <- function(lista) {
  vp <- variancia_P(data$tempo)
  dp <- sqrt(vp)
  return(dp)
}

round(dp(data$tempo))
```

```r
#Desvio Padrão Amostral
dpa <- function(lista) {
  va <- variancia_A(data$tempo)
  dpa <- sqrt(va)
  return(dpa)
}

round(dpa(data$tempo))
```

```r
#Coefiente De Variação Amostral:
cv <- function(lista) {
  cv <- (dp(data$tempo)/mean(data$tempo))*100
  return(cv)
}
round(cv(data$tempo))
```

```r
#Coefiente De Variação Amostral:
cva <- function(lista) {
  cva <- (dpa(data$tempo)/mean(data$tempo))*100
  return(cva)
}
round(cva(data$tempo))
```

```r
#Quartil 1:
posicao_mediana <- (length(data$tempo)+1)/2
ordem <- sort(data$tempo)
metade1 <- ordem[0:posicao_mediana]
q1 <- median(metade1)
round(q1)
```
